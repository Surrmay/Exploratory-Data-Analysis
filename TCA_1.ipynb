```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('Customer Churn.csv')
df
```

```python
df.info()
```

CHANGED THE NULL VALUES OF TOTAL CHARGES TO 0 AND REPLACED THE DATATYPE WITH FLOAT.

```python
df['TotalCharges'] = df['TotalCharges'].replace(' ','0')
df['TotalCharges'] = df['TotalCharges'].astype('float')
```

CONVERTED ALL THE 0 AND 1 VALUES OF SENIOR CITIZEN TO YES OR NO

```python
def conv(value):
    if value == 1:
        return 'yes'
    else:
        return 'no'

df['SeniorCitizen'] = df['SeniorCitizen'].apply(conv)
```

```python
ax = sns.countplot(x = 'Churn', data = df)
ax.bar_label(ax.containers[0])
plt.title('Count of costumers churn')
plt.show()
```

FROM THE PIECHART WE CAN CONCLUDE THE 26% OF OUR CONSTUMERS HAVE CHURNED OUT

```python
plt.figure(figsize=(3,4))
gb = df.groupby('Churn').agg({'Churn':"count"})
plt.pie(gb['Churn'],labels=gb.index,autopct='%1.2f%%')
plt.title('Percentage of Constumers churn')
plt.show()
```

EXPLORE THE REASON FOR CHURNING OF CUSTOMERS

```python
sns.countplot(x = 'gender', data = df, hue = 'Churn')
plt.show()
```

```python
plt.figure(figsize=(4,4))
ax = sns.countplot(x = 'SeniorCitizen', data = df)
ax.bar_label(ax.containers[0])
plt.title('Churn by SeniorCitizen')
plt.show()
```

```python

counts = pd.crosstab(df['SeniorCitizen'], df['Churn'])

# Convert counts to percentages
counts_percent = counts.div(counts.sum(axis=1), axis=0) * 100

# Get the column names to avoid KeyError
churn_labels = counts_percent.columns

# Plot stacked bar chart
fig, ax = plt.subplots(figsize=(4, 4))

# Plot bars
ax.bar(counts_percent.index, counts_percent[churn_labels[0]], label=f'({churn_labels[0]})', color='#1f77b4')
ax.bar(counts_percent.index, counts_percent[churn_labels[1]], bottom=counts_percent[churn_labels[0]], label=f'{churn_labels[1]})', color='#ff7f0e')

# Add percentage labels
for i in range(len(counts_percent)):
    plt.text(i, counts_percent[churn_labels[0]][i] / 2, f'{counts_percent[churn_labels[0]][i]:.1f}%', ha='center', color='white')
    plt.text(i, counts_percent[churn_labels[0]][i] + counts_percent[churn_labels[1]][i] / 2, f'{counts_percent[churn_labels[1]][i]:.1f}%', ha='center', color='white')

# Set title and labels
plt.title('Churn by SeniorCitizen (Percentage)')
plt.xlabel('SeniorCitizen')
plt.ylabel('Percentage (%)')
plt.legend(title='Churn', loc='upper left', bbox_to_anchor=(1, 1))
plt.tight_layout()
plt.show()
```

FROM THE ABOVE ANALYSIS, WE CAN SEE THAT COMPARATIVELY A GREATER PERCENTAGE OF PEOPLE IN SENIOR CITIZEN CATEGORY HAVE CHURNED

PEOPLE WHO HAVE USED OUR SERVICES FOR LONGER DURATION OF TIME HAVE STAYED AND WHO HAVE USED IT FOR A SHORTER DURATION HAVE CHURNED

```python
plt.figure(figsize=(4,4))
ax = sns.countplot(x = 'Contract', data = df,hue = 'Churn')
ax.bar_label(ax.containers[0])
plt.title("Count of Customer by Contract")
plt.show()
```

PEOPLE WHO HAVE MONTH TO MONTH CONTRACTS ARE TEND TO CHURN WHEREAS COSTUMERS WITH A YEAR OR TWO YEAR CONTRACTS ARE TEND TO STAY

```python
# List of columns to plot
columns = [
    'PhoneService', 'MultipleLines', 'InternetService', 'OnlineSecurity', 
    'OnlineBackup', 'DeviceProtection', 'TechSupport', 'StreamingTV', 'StreamingMovies'
]

# Number of columns and rows for subplots
n_cols = 3
n_rows = (len(columns) + n_cols - 1) // n_cols  # Calculate rows needed based on number of plots

# Create a figure and a set of subplots
fig, axes = plt.subplots(n_rows, n_cols, figsize=(15, 10))  # Adjust figure size as needed
fig.suptitle('Count Plots for Each Service', fontsize=16)

# Flatten axes array for easy iteration
axes = axes.flatten()

# Loop over each column and plot in the corresponding subplot
for i, col in enumerate(columns):
    sns.countplot(x=col, data=df, ax=axes[i],hue = df['Churn'])
    axes[i].set_title(f'{col} Count')
    axes[i].set_xlabel('')
    axes[i].set_ylabel('Count')

# Remove any empty subplots if the grid is larger than needed
for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout(rect=[0, 0, 1, 0.96])  # Adjust layout to fit title
plt.show()
```
